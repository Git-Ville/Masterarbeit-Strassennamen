# Automatische Kategorisierung lokaler Bezug und Wirkungsbereich

Dieses Skript vergibt automatisch die Werte in den Kategorien "lokaler Bezug" und Wirkungsbereich.

```python
# benötigte Bibliotheken
import pandas as pd
```

```python
# CSV-Datei einlesen
file_path_persons = '../../data/processed/Data_Masterarbeit - list_named_after_humans.csv'
file_path_all_streets = '../../data/processed/Data_Masterarbeit - list_complete.csv'
file_path_street_layout = "../../data/raw/lhd_strassenzuege.csv"

df_persons = pd.read_csv(file_path_persons, sep=",")
df_all = pd.read_csv(file_path_all_streets, sep=",")
df_streets = pd.read_csv(file_path_street_layout, sep=";")
```

```python
# Liste der relevanten Begriffe für Dresden und Stadtteile
dresden_Begriffe = [
    "Dresden", "Altstadt", "Blasewitz", "Cotta", "Klotzsche", "Leuben",
    "Loschwitz", "Neustadt", "Pieschen", "Plauen", "Prohlis",
    "Friedrichstadt", "Johannstadt", "Südvorstadt", "Striesen", "Gruna", "Laubegast",
    "Strehlen", "Leubnitz-Neuostra", "Niedersedlitz", "Tolkewitz", "Kleinpestitz",
    "Löbtau", "Gorbitz", "Räcknitz", "Coschütz", "Pennrich", "Hellerau",
    "Weißig", "Pillnitz", "Kaditz", "Übigau", "Omsewitz", "Mobschatz",
    "Wilschdorf", "Reick", "Prohlis-Nord", "Prohlis-Süd", "Gittersee", "Flurname", "Stadtteil"
]
# Funktion zum Prüfen, ob einer der Begriffe in den relevanten Spalten vorkommt für alle Straßennamen
def hat_lokalen_bezug_all(row):
    for spalte in ["Straßenname", "benannt nach", "Benennungstyp" ]:
        if any(stadtteil in str(row[spalte]) for stadtteil in dresden_Begriffe):
            return "ja"
    return "nein"

# Funktion zum Prüfen, ob einer der Begriffe in den relevanten Spalten vorkommt für personenbezogene Straßennamen
def hat_lokalen_bezug_persons(row):
    for spalte in ["Straßenname", "benannt nach", "Geburtsort", "Sterbeort" ]: 
        if any(stadtteil in str(row[spalte]) for stadtteil in dresden_Begriffe):
            return "ja"
    return "nein"

# Neue Spalte hinzufügen
df_all["lokaler Bezug"] = df_all.apply(hat_lokalen_bezug_all, axis=1)
df_persons["lokaler Bezug"] = df_persons.apply(hat_lokalen_bezug_persons, axis=1)

# Datei speichern
df_all.to_csv("../../data/processed/strassennamen_gesamt_mit_lokalem_Bezug.csv", index=False)
df_persons.to_csv("../../data/processed/strassennamen_personen_mit_lokalem_Bezug.csv", index=False)
```

```python
# CSV-Datei einlesen
datei_lokaler_Bezug = "../../data/processed/strassennamen_personen_mit_lokalem_Bezug.csv"  # Dateipfad anpassen
df_lokaler_Bezug = pd.read_csv(datei_lokaler_Bezug)

# Wirkungsbereich setzen, wenn lokaler Bezug 'ja'
df_lokaler_Bezug.loc[(df_lokaler_Bezug["lokaler Bezug"] == "ja"), "Wirkungsbereich"] = "lokal"

# Aktualisierte Datei speichern
df_lokaler_Bezug.to_csv("../../data/processed/strassennamen_mit_wirkungsbereich.csv", index=False)
```
