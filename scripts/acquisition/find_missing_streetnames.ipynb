# Erhebung Geodaten Orte
Dieses Skript vergleicht die Datenbank der Masterarbeit mit dem offiziellen Dresdner Straßenverzeichnis und durchsucht diese nach Straßen, die in einer der beiden Verzeichnisse fehlen.

```python
# Benötigte Bibliotheken laden
import pandas as pd
```

```python

# Funktion definieren, welche die Straßennamen normalisiert
def normalize_streetname(name):
    if pd.isna(name):
        return None
    return (name
            .strip()                # Leerzeichen vorne/hinten weg
            .replace("ß", "ss")     # ß und ss vereinheitlichen
            .replace(".", "")       # Abkürzungspunkte entfernen
            .lower())               # alles klein schreiben

# Daten einlesen
my_list = pd.read_csv("../../data/processed/Data_Masterarbeit - list_complete.csv", delimiter=",", on_bad_lines='skip')
opendata_dresden_list = pd.read_csv("../../Data/raw/opendata_strassenverzeichnis_dresden_2025.csv", delimiter=";", on_bad_lines='skip')

# Normalisierte Sets
streets_my_list = set(my_list["Straßenname"].dropna().map(normalize_streetname).unique())
streets_opendata_dresden_list = set(opendata_dresden_list["Straßenname"].dropna().map(normalize_streetname).unique())

# Mengenoperationen
missing_values_mylist = streets_opendata_dresden_list - streets_my_list
unique_values_mylist = streets_my_list - streets_opendata_dresden_list

# Ausgabe
print("Folgende Straßen fehlen in meiner Liste:", missing_values_mylist)
print("Folgende Straßen sind fälschlicherweise in meiner Liste:", unique_values_mylist)
```
